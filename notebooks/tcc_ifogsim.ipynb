```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
xAxis = ['4x4', '4x8', '4x12']
columns = ['Elapsed time','CPU (%)','Real (MB)','Virtual (MB)']
```

# Carregando dados de log gerados pelo psrecord

O psrecord gera logs de uso de cpu, uso de memoria e também plots comparativos


```python
#edge
e_4_4 = pd.read_csv('/content/drive/MyDrive/tcc-dados/ifogsim/edge/4x4e.txt', sep="\t",skiprows = 1,header = None)
e_4_8 = pd.read_csv('/content/drive/MyDrive/tcc-dados/ifogsim/edge/4x8e.txt', sep="\t",skiprows = 1,header = None)
e_4_12 = pd.read_csv('/content/drive/MyDrive/tcc-dados/ifogsim/edge/4x12e.txt', sep="\t",skiprows = 1,header = None)
```

```python
#cloud
c_4_4 = pd.read_csv('/content/drive/MyDrive/tcc-dados/ifogsim/cloud/4x4c.txt', sep="\t",skiprows = 1,header = None)
c_4_8 = pd.read_csv('/content/drive/MyDrive/tcc-dados/ifogsim/cloud/4x8c.txt', sep="\t",skiprows = 1,header = None)
c_4_12 = pd.read_csv('/content/drive/MyDrive/tcc-dados/ifogsim/cloud/4x12c.txt', sep="\t",skiprows = 1,header = None)
```

```python
# clean white spaces and return series

def clean_white_spaces_dataframe(df):
  aux = df[0].map(lambda x: x.replace(' ', ',').split(','))
  res = aux.map(lambda x : list(filter(lambda y: y != "",x)))
  return res
```

```python
# create array for dataframe

def create_array_by_series(sr):
  arr = []
  for obj in sr:
    arr.append(obj)
  return arr
```

# Pré-processamento dos dados

nesse passo foi retirado os espaços em branco que tinham no arquivo original e com isso foram gerados dataframes do pandas. Com o auxílio de algumas funções da própria biblioteca do pandas foram pegados os máximo valores no uso de Cpu.

```python
# edge transformations

t1 = clean_white_spaces_dataframe(e_4_4)
t2 = create_array_by_series(t1)

ne_4_4 = pd.DataFrame(t2, columns=columns)

######################################

t3 = clean_white_spaces_dataframe(e_4_8)
t4 = create_array_by_series(t3)

ne_4_8 = pd.DataFrame(t4, columns=columns)

######################################

t5 = clean_white_spaces_dataframe(e_4_12)
t6 = create_array_by_series(t5)

ne_4_12 = pd.DataFrame(t6, columns=columns)
```

```python
# cloud transformations

t1 = clean_white_spaces_dataframe(c_4_4)
t2 = create_array_by_series(t1)

nc_4_4 = pd.DataFrame(t2, columns=columns)

######################################

t3 = clean_white_spaces_dataframe(c_4_8)
t4 = create_array_by_series(t3)

nc_4_8 = pd.DataFrame(t4, columns=columns)

######################################

t5 = clean_white_spaces_dataframe(c_4_12)
t6 = create_array_by_series(t5)

nc_4_12 = pd.DataFrame(t6, columns=columns)
```

```python
# get cpu values from dataframes

max_edge_cpu_values = []
max_edge_mem_values = []
```

```python
# get memory values from dataframes

max_cloud_cpu_values = []
max_cloud_mem_values = []
```

```python
# get execution time ms

max_time_edge = [2200,3852,6848] # 4x4 4x8 4x12

max_time_cloud = [3768,8955,20249] # 4x4 4x8 4x12
```

```python
# get network usage 

max_net_edge = [5876.8,17116.8,33718.64]

max_net_cloud = [212390.44,445765.8,585253.48]
```

```python
# Edge max values
e_cpu_max_4_4 = ne_4_4['CPU (%)'].map(lambda x: float(x))
e_mem_max_4_4 = ne_4_4['Real (MB)'].map(lambda x: float(x))
max_edge_cpu_values.append(e_cpu_max_4_4.max())
max_edge_mem_values.append(e_mem_max_4_4.max())

e_cpu_max_4_8 = ne_4_8['CPU (%)'].map(lambda x: float(x))
e_mem_max_4_8 = ne_4_8['Real (MB)'].map(lambda x: float(x))
max_edge_cpu_values.append(e_cpu_max_4_8.max())
max_edge_mem_values.append(e_mem_max_4_8.max())

e_cpu_max_4_12 = ne_4_12['CPU (%)'].map(lambda x: float(x))
e_mem_max_4_12 = ne_4_12['Real (MB)'].map(lambda x: float(x))
max_edge_cpu_values.append(e_cpu_max_4_12.max())
max_edge_mem_values.append(e_mem_max_4_12.max())
```

```python
# Cloud max values
c_cpu_max_4_4 = nc_4_4['CPU (%)'].map(lambda x: float(x))
c_mem_max_4_4 = nc_4_4['Real (MB)'].map(lambda x: float(x))
max_cloud_cpu_values.append(c_cpu_max_4_4.max())
max_cloud_mem_values.append(c_mem_max_4_4.max())

c_cpu_max_4_8 = nc_4_8['CPU (%)'].map(lambda x: float(x))
c_mem_max_4_8 = nc_4_8['Real (MB)'].map(lambda x: float(x))
max_cloud_cpu_values.append(c_cpu_max_4_8.max())
max_cloud_mem_values.append(c_mem_max_4_8.max())

c_cpu_max_4_12 = nc_4_12['CPU (%)'].map(lambda x: float(x))
c_mem_max_4_12 = nc_4_12['Real (MB)'].map(lambda x: float(x))
max_cloud_cpu_values.append(c_cpu_max_4_12.max())
max_cloud_mem_values.append(c_mem_max_4_12.max())
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

## Gráfico do uso de CPU da máquina que está rodando a simulação

```python
ed_44 = ne_4_4['CPU (%)'].map(lambda x: float(x)).map(lambda x: round(x)).drop_duplicates()/100
ed_48 = ne_4_8['CPU (%)'].map(lambda x: float(x)).map(lambda x: round(x)).drop_duplicates()/100
ed_412 = ne_4_12['CPU (%)'].map(lambda x: float(x)).map(lambda x: round(x)).drop_duplicates()/100

cl_44 = nc_4_4['CPU (%)'].map(lambda x: float(x)).map(lambda x: round(x)).drop_duplicates()/100
cl_48 = nc_4_8['CPU (%)'].map(lambda x: float(x)).map(lambda x: round(x)).drop_duplicates()/100
cl_412 = nc_4_12['CPU (%)'].map(lambda x: float(x)).map(lambda x: round(x)).drop_duplicates()/100



fig = make_subplots(rows=1, cols=3,  subplot_titles=('4 x 4 sensores','4 x 8 sensores','4 x 12 sensores'))

fig.add_trace(go.Box(y=ed_44,name='edge'), row=1,col=1)
fig.add_trace(go.Box(y=cl_44,name='cloud'),row=1,col=1)

fig.add_trace(go.Box(y=ed_48,name='edge'), row=1,col=2)
fig.add_trace(go.Box(y=cl_48,name='cloud'),row=1,col=2)

fig.add_trace(go.Box(y=ed_412,name='edge'), row=1,col=3)
fig.add_trace(go.Box(y=cl_412,name='cloud'),row=1,col=3)

fig.update_layout(
    title="Cloud x Edge - CPU",
    yaxis_title="CPU(%)",
    showlegend=False,
     width=1200,
    height=500,
     font=dict(
        family="Courier New, monospace",
        size=18,
        color="RebeccaPurple"
    )
)

fig.show()
```

## Gráfico do uso de Memória da máquina que está rodando a simulação

```python
ed_44 = ne_4_4['Real (MB)'].map(lambda x: float(x)).map(lambda x: round(x)).drop_duplicates()
ed_48 = ne_4_8['Real (MB)'].map(lambda x: float(x)).map(lambda x: round(x)).drop_duplicates()
ed_412 = ne_4_12['Real (MB)'].map(lambda x: float(x)).map(lambda x: round(x)).drop_duplicates()

cl_44 = nc_4_4['Real (MB)'].map(lambda x: float(x)).map(lambda x: round(x)).drop_duplicates()
cl_48 = nc_4_8['Real (MB)'].map(lambda x: float(x)).map(lambda x: round(x)).drop_duplicates()
cl_412 = nc_4_12['Real (MB)'].map(lambda x: float(x)).map(lambda x: round(x)).drop_duplicates()



fig = make_subplots(rows=1, cols=3,  subplot_titles=('4 x 4 sensores','4 x 8 sensores','4 x 12 sensores'))

fig.add_trace(go.Box(y=ed_44,name='edge'), row=1,col=1)
fig.add_trace(go.Box(y=cl_44,name='cloud'),row=1,col=1)

fig.add_trace(go.Box(y=ed_48,name='edge'), row=1,col=2)
fig.add_trace(go.Box(y=cl_48,name='cloud'),row=1,col=2)

fig.add_trace(go.Box(y=ed_412,name='edge'), row=1,col=3)
fig.add_trace(go.Box(y=cl_412,name='cloud'),row=1,col=3)

fig.update_layout(
    title="Cloud x Edge - Mem",
    yaxis_title="Memory (mb)",
    showlegend=False
)

fig.show()
```

```python
e_44_mean = e_mem_max_4_4.mean()
e_48_mean = e_mem_max_4_8.mean()
e_412_mean = e_mem_max_4_12.mean()

e_memo = [round(e_44_mean,2) ,round(e_48_mean,2),round(e_412_mean,2)]


c_44_mean = c_mem_max_4_4.mean()
c_48_mean = c_mem_max_4_8.mean()
c_412_mean = c_mem_max_4_12.mean()

c_memo = [round(c_44_mean,2),round(c_48_mean,2),round(c_412_mean,2)]

fig = go.Figure(data=[
    go.Bar(name='Edge', x=xAxis, y=e_memo,text=e_memo, textposition='auto',marker_color='green'),
    go.Bar(name='Cloud', x=xAxis, y=c_memo, text=c_memo, textposition='auto',marker_color='firebrick')
])

# Change the bar mode
fig.update_layout(title='Cloud x Edge - mem',xaxis_title='type',yaxis_title="Memory (mb)", barmode='group')
fig.show()

```
