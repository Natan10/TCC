```python
import pandas as pd
from datetime import timedelta, datetime
import random
```

```python
df_node_1 = pd.DataFrame([], columns=['Oximetria','Hora','Data'])
df_node_2 = pd.DataFrame([], columns=['Oximetria','Hora','Data'])
```

```python
# gerando data e hora

day = datetime.fromisoformat('2022-06-16T00:00:00')
oximetro_days = []
oximetro_hour = []

new_date = day + timedelta(minutes=1)

for i in range(0,10080):
  oximetro_days.append(new_date.strftime('%d-%m-%y'))
  oximetro_hour.append(new_date.strftime('%H:%M:%S'))
  new_date = new_date + timedelta(minutes=1)

oximetro_days[10079] = '22-06-22'
```

```python
# gerando dados do ox√≠metro
random.seed(100)

oximetro_data_1 = [random.randint(40,98) for _ in range(0,10080)]
oximetro_data_2 = [random.randint(90,98) for _ in range(0,10080)]
```

```python
df_node_1['Hora'] = oximetro_hour
df_node_1['Data'] = oximetro_days
df_node_2['Hora'] = oximetro_hour
df_node_2['Data'] = oximetro_days

df_node_1['Oximetria'] = oximetro_data_1
df_node_2['Oximetria'] = oximetro_data_2
```
