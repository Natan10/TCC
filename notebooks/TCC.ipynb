# Testes de ações nas instâncias

## Stress

### Cloud

O Fogify injeta uma carga de trabalho extra em nós específicos para simular as interferências do serviço colocado em um nó. Isso pode ser utilizado para emular uma situação em que a instância, no caso a cloud teria que realizar um processamento muito grande de dados, podemos ver a diferença disso nos gráficos. Isso pode ser extendido aos outros nós, como os nó de fog que podem ter que realizar algum processamento que demande muito da sua cpu.

```python
# Antes de aplicar o stress na instância de cpu

cloud_cpu_values = fogify.get_metrics_from('cloud-server.1')

plt.figure(figsize=(15,8))

sns.lineplot(x=cloud_cpu_values.index.values,y=cloud_cpu_values['cpu_util']).set(title='Cloud before stress')
```

```python
fogify.stress('cloud-server.1',duration=100,cpu=50)
```

```python
# depois de aplicar o stress na instância de cpu

cloud_cpu_values = fogify.get_metrics_from('cloud-server.1')

plt.figure(figsize=(15,8))

sns.lineplot(x=cloud_cpu_values.index.values,y=cloud_cpu_values['cpu_util']).set(title='Cloud after stress')
```

## Horizontal_scaling_down

```python
fogify.horizontal_scaling_down('oxi-workload-1')
```

Vemos na imagem abaixo que a instancia que representava o oxímetro 1 não esta mais ativa no cluster do docker swarm, isso pois aplicamos o horizontal scaling down. Isso pode ser usado para emular situações em que o serviço deseja apresenta algum defeito ou simplesmente deixa de funcionar.

## Horizontal_scaling_up

```python
fogify.horizontal_scaling_up('oxi-workload-1')
```

Da mesma forma que fazemos o scaling down, o fogify também permite fazer o horizontal scaling up para adicionar mais instâncias ao cenário emulado. Isso ajuda, caso os desenvolvedores queiram adicionar mais máquinas, sejam fog nodes, instâncias de sensores e etc.
